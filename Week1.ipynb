```python
def find_number(line_span):
    num_list = [char for char in line_span.split(".") if char != ""]
    new_list = []
    for num in num_list:
        for char in num:
            if not char.isdigit():
                num = num.replace(char, "?")

        new_list.extend([int(n) for n in num.split("?") if n != ""])
    return new_list
```
